```python
from __future__ import print_function

import numpy as np 
import pandas as pd

from sklearn.model_selection import train_test_split
from LinearRegression import LinearRegression

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("data.csv")
X = df['X'].values
y = df['y'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.22, random_state=42)
```

```python
m, b, cost = LinearRegression(X_train, y_train, activation="gradient_descent")
print("The slope is: {}".format(m))
print("The bias is: {}".format(b))
print("The cost is: {}".format(cost))

plt.figure(figsize=(13,7))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title("Gradient Descent", fontsize=30)

plt.scatter(X_test, y_test, label="test points", color="blue")
plt.plot(X_test, m*X_test + b, label="trained line on test", color="black")

plt.legend(loc="lower right", fontsize=20)
```

```python
m, b, cost = LinearRegression(X_train, y_train, activation="sgd", batch_size=10)
print("The slope is: {}".format(m))
print("The bias is: {}".format(b))
print("The cost is: {}".format(cost))

plt.figure(figsize=(13,7))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title("Stochastic Gradient Descent", fontsize=30)

plt.scatter(X_test, y_test, label="test points", color="blue")
plt.plot(X_test, m*X_test + b, label="trained line on test", color="black")

plt.legend(loc="lower right", fontsize=20)
```

```python
m, b, cost = LinearRegression(X_train, y_train, activation="momentum", batch_size=10)
print("The slope is: {}".format(m))
print("The bias is: {}".format(b))
print("The cost is: {}".format(cost))

plt.figure(figsize=(13,7))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title("Momentum", fontsize=30)

plt.scatter(X_test, y_test, label="test points", color="blue")
plt.plot(X_test, m*X_test + b, label="trained line on test", color="black")

plt.legend(loc="lower right", fontsize=20)
```

```python
m, b, cost = LinearRegression(X_train, y_train, activation="nesterov", batch_size=10)
print("The slope is: {}".format(m))
print("The bias is: {}".format(b))
print("The cost is: {}".format(cost))

plt.figure(figsize=(13,7))
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.title("Nesterov Accelerated Gradient", fontsize=30)

plt.scatter(X_test, y_test, label="test points", color="blue")
plt.plot(X_test, m*X_test + b, label="trained line on test", color="black")

plt.legend(loc="lower right", fontsize=20)
```
